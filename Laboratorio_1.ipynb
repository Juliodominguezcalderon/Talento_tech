```python
#importar la librería de panda
import pandas as pd
```

```python
#Asignarle a la variable df_main (Dataframe principal) la información del archivo
df_main = pd.read_csv("/content/dataset_banco.csv",delimiter=",")
#Visualizar los datos para validar que hayan cargado
print(df_main.head(5))
```

```python
#Conocer los tipo de datos de las columnas
df_main.dtypes
```

```python
#Crear una copia fidelidad de dataframe cargado para iniciar las transformaciones
df_copy=df_main.copy()
```

```python
#contar la cantidad de filas y columnas del dataframe
print(df_copy.shape)
print(df_main.shape)
```

```python
#contar la cantidad de filas por columnas del dataframe
df_copy.count()
```

```python
df_copy.head(5)
```

```python
#Renombrar el nombre de las columnas del dataframe
df_copy = df_copy.rename(columns={
                              'age':'edad',
                              'job':'trabajo',
                              'marital':'estado civil',
                              'education':'educacion',
                              'default':'morosidad',
                              'balance':'saldo',
                              'housing':'vivienda',
                              'loan':'prestamo',
                              'contact':'contacto',
                              'day':'dia',
                              'month':'mes',
                              'duration':'duracion',
                              'campaign':'campaña',
                              'pdays':'dias_previos',
                              'previous':'cantidad de contactos anteriores',
                              'poutcome':'resultado',
                              'y':'respuesta'
                              })
#verificamos los cambios
print(df_copy.head())
```

```python
df_copy.head(50)
```

```python
#Modo 2 de renombrar las columnas
df_copy2 = df_main.copy()
df_copy2.columns = ['edad',
              'trabajo',
              'estado_civil',
              'educacion',
              'incumplimiento',
              'saldo',
              'vivienda',
              'prestamo',
              'contacto',
              'dia',
              'mes',
              'duracion',
              'campaña',
              'dias_previos',
              'anterior',
              'resultado_prev',
              'respuesta']
print(df_copy2.columns)
```

```python
#Visualizar las variables categóticas y númericas
df_copy.info()
```

```python
#clasificar columnas
categoricas = df_copy.select_dtypes(include=['object']).columns.tolist()
numericas = df_copy.select_dtypes(include=['int64', 'float64']).columns.tolist()
print("Columnas categóricas:", categoricas)
print("Columnas numéricas:", numericas)
```
