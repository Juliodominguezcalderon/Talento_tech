```python
#Identificamos los datos "Errones" para el campo edad
df_edad_100 = df_copy[df_copy['edad'] > 100]
df_edad_100
```

```python
#Generamos un data frame con datos limpios eliminando las filas identificadas el df_edad_100 mayor a es 100 años
df_Datos_limpios = df_copy.drop(df_edad_100.index)
df_Datos_limpios
```

```python
#Generamos un dataframe para los postenciales clientes que estan interesados en el producto
#para que se active la respectiva campaña
df_yes = df_Datos_limpios[df_Datos_limpios['respuesta'] == 'yes']
df_yes.to_csv('df_yes.csv', index=False)
print(f"Data frame principal (Filas, Columas): {df_main.shape}")
print(f"Data frame copia del principal (Filas, Columas): {df_copy.shape}")
print(f"Data frame principal (Filas, Columas): {df_Datos_limpios.shape}")
print(f"Data frame personas interesadas (Filas, Columas): {df_yes.shape}")
```

```python
# prompt: Crear una columna de rango de edad

# Definir los rangos de edad
bins = [0, 18, 30, 45, 60, 100]
labels = ['0-18', '19-30', '31-45', '46-60', '61+']

# Crear la columna de rango de edad
df_Datos_limpios['rango_edad'] = pd.cut(df_Datos_limpios['edad'], bins=bins, labels=labels, right=False)

# Mostrar el DataFrame con la nueva columna
print(df_Datos_limpios.head())
```

```python
# Concoer la distirbución de la morosidad y el sado por rango de edad

import matplotlib.pyplot as plt
import seaborn as sns

# Agrupar por rango de edad y calcular la proporción de morosidad y el saldo promedio
morosidad_por_edad = df_Datos_limpios.groupby('rango_edad')['morosidad'].value_counts()
saldo_por_edad = df_Datos_limpios.groupby('rango_edad')['saldo'].mean()

# Crear el gráfico de barras para la morosidad
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
morosidad_por_edad.plot(kind='bar', stacked=True)
plt.title('Proporción de Morosidad por Rango de Edad')
plt.xlabel('Rango de Edad')
plt.ylabel('Proporción')
plt.legend(title='Morosidad')

# Crear el gráfico de barras para el saldo promedio
plt.subplot(1, 2, 2)
saldo_por_edad.plot(kind='bar')
plt.title('Saldo Promedio por Rango de Edad')
plt.xlabel('Rango de Edad')
plt.ylabel('Saldo Promedio')

plt.tight_layout()
plt.show()
```

```python
# Visualizar los valores únicos del campo 'trabajo'
print(df_Datos_limpios['trabajo'].unique())
# Contar la frecuencia de cada valor en el campo 'trabajo'
print(df_Datos_limpios['trabajo'].value_counts())
```

Como podemos ver el campo Trabajo contiene información relacionada pero escrita de forma distinta, por ejemplo: para management existen management  9454, Management 2 y MANAGEMENT 2.
Lo anterior no corresponde a Trabajos distintos sino al mismo cargo y debemos corregir esto para continuar con el análisis.



```python
# Convertir los valores del campo 'trabajo' a minúsculas esto ayuda a unificar la información escrita de forma distinta
df_Datos_limpios['trabajo'] = df_Datos_limpios['trabajo'].str.lower()

#Count the occurrences of 'management' in the 'trabajo' column
management_count = df_Datos_limpios['trabajo'][df_Datos_limpios['trabajo'] == 'management'].count()

# Print the count
print(management_count)
```

```python
#Verificar los cambios
print(df_Datos_limpios['trabajo'].unique())
print(df_Datos_limpios['trabajo'].value_counts())
```

```python
#Los trabajos admin y administrative son los mismos

# Reemplazar 'admin.' por 'administrative' en la columna 'trabajo'
df_Datos_limpios['trabajo'] = df_Datos_limpios['trabajo'].replace('admin.', 'administrative')

# Verificar los cambios
print(df_Datos_limpios['trabajo'].unique())
print(df_Datos_limpios['trabajo'].value_counts())
```

```python
# Generamos el saldo promedio por trabajo
saldo_promedio_por_trabajo = df_Datos_limpios.groupby('trabajo')['saldo'].mean()
plt.figure(figsize=(10, 6))
saldo_promedio_por_trabajo.plot(kind='bar')
plt.title('Saldo Promedio por Trabajo')
plt.xlabel('Trabajo')
plt.ylabel('Saldo Promedio')
plt.xticks(rotation=90) #Rotar las etiquetas del eje x para mejor legibilidad
plt.tight_layout()
plt.show()
```

```python
morosidad_por_trabajo = df_Datos_limpios.groupby('trabajo')['morosidad'].value_counts().unstack()
plt.figure(figsize=(10, 6))
morosidad_por_trabajo.plot(kind='bar')
plt.title('Saldo Promedio por Trabajo')
plt.xlabel('Trabajo')
plt.ylabel('Saldo Promedio')
plt.xticks(rotation=90) #Rotar las etiquetas del eje x para mejor legibilidad
plt.tight_layout()
plt.show()
```

```python
estado_civil_counts = df_Datos_limpios['estado civil'].value_counts()
print(estado_civil_counts)
```

```python
#Corregimos el campo estado civil para unificar valores mal escritos
df_Datos_limpios['estado civil'] = df_Datos_limpios['estado civil'].str.lower() #unificamos todo en minúsculas
df_Datos_limpios['estado civil'] = df_Datos_limpios['estado civil'].replace('div.', 'divorced')#cambiamos div. por divorced
print(df_Datos_limpios['estado civil'].unique())
```

```python
saldo_promedio_por_estado_civil = df_Datos_limpios.groupby('estado civil')['saldo'].mean().
plt.figure(figsize=(10, 6))
saldo_promedio_por_estado_civil .plot(kind='bar')
plt.title('Saldo promedio por estado civil')
plt.xlabel('Trabajo')
plt.ylabel('Saldo Promedio')
plt.xticks(rotation=90) #Rotar las etiquetas del eje x para mejor legibilidad
plt.tight_layout()
plt.show()
```

```python
# Agrupar por estado civil y respuesta, y contar la cantidad de registros
respuesta_por_estado_civil = df_Datos_limpios.groupby(['estado civil', 'respuesta'])['respuesta'].count().unstack()

# Crear el gráfico de barras
respuesta_por_estado_civil.plot(kind='bar', figsize=(10, 6))
plt.title('Respuesta por Estado Civil')
plt.xlabel('Estado Civil')
plt.ylabel('Cantidad')
plt.xticks(rotation=0)
plt.legend(title='Respuesta')
plt.show()
```

```python
# Agrupar por estado civil y respuesta, y contar la cantidad de registros
respuesta_por_trabajo= df_Datos_limpios.groupby(['trabajo', 'respuesta'])['respuesta'].count().unstack()

# Crear el gráfico de barras
respuesta_por_trabajo.plot(kind='bar', figsize=(10, 6))
plt.title('Respuesta por trabajo')
plt.xlabel('Trabajo')
plt.ylabel('Cantidad')
plt.xticks(rotation=90)
plt.legend(title='Respuesta')
plt.show()
```
